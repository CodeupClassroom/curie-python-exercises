```python
students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan', 'Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=40, high=90, size=len(students))
english_grades = np.random.randint(low=40, high=90, size=len(students))
reading_grades = np.random.randint(low=40, high=90, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades,
                   'classroom': np.random.choice(['A', 'B'], len(students))})
df
```

```python
df.groupby("classroom").reading.median()
```

```python
df.groupby("classroom").reading.describe()
```

```python
df.groupby("classroom").reading.mean()
```

```python
df.classroom.value_counts()
```

```python
# possible to iterate a groupby object (not necessarily recommended)
for x in df.groupby("classroom"):
    print(x[1])
    break
```

```python
classroom_a = df[df.classroom == "A"]
classroom_a
```

```python
df.groupby("classroom").math.mean()
```

```python
df.groupby("classroom").english.max()
```

```python
df.groupby("classroom").agg(["std", "median"])
```

```python
df.groupby("classroom").agg(["std", "median"]).reading
```

```python
df["passing_math"] = df.math >= 70
df["passing_english"] = df.english >= 70
df["passing_reading"] = df.reading >= 70
```

```python
df
```

```python
df.groupby(["passing_english", "passing_math", "passing_reading"]).mean()
```

```python
passing_all = df.passing_english & df.passing_reading & df.passing_math
```

```python
df["passing_all"] = passing_all
```

```python
df
```

```python
df.groupby("passing_all").math.min()
```

```python
df.groupby("passing_all").math.mean()
```
