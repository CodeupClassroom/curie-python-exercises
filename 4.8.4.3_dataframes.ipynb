## 1. Create student grades DataFrame object

```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

### Peek at DataFrame

```python
df.head()
```

```python
# use shape attribute

df.shape
```

```python
# use info method to view both datatypes and potential missing values

df.info()
```

```python
# use describe method to produce descriptive statistics for columns with numeric datatypes

df.describe()
```

### a. Create a column named `passing_english` that indicates whether each student has a passing grade in reading.

```python
# create a Series of boolean values or a boolean mask

df.english >= 70
```

```python
# assign the series to a new column we create in our DataFrame

df['passing_english'] = df.english >= 70
```

```python
# check dataframe for new column

df.head()
```

```python
# How many people are passing English? Use sum() function to add the True bool values

df['passing_english'].sum()
```

```python
# How many students are failing English? Use .sum() method

(df['passing_english'] == False).sum()
```

```python
# View both 

df['passing_english'].value_counts()
```

### b. Sort the english grades by the `passing_english` column. How are duplicates handled?

- `.sort_values` returns a sorted copy of a given DataFrame unless inplace=True

- It looks like duplicate values are handled according to the index value, small to large or ascending.

```python
# DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last')
# sorts all of the rows in the DF using the column passed

df.sort_values(by='passing_english')
```

### c. Sort the english grades first by `passing_english` and then by student `name`. 

All the students that are failing english should be first, and within the students that are failing english they should be ordered alphabetically. The same should be true for the students passing english. 

<div class="alert alert-block alert-info">
<b>Hint:</b> You can pass a list to the .sort_values method.
</div>

```python
# Now we see that Alan comes before Albert because there is a secondary sort going on alphabetically by name

df.sort_values(by=['passing_english', 'name'])
```

### d. Sort the english grades first by `passing_english`, and then by the actual `english` grade, similar to how we did in the last step.

```python
df.sort_values(by=['passing_english', 'english'])
```

### e. Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.

```python
# just calling the .sum() method has a default of axis=0 which sums the rows for each column

df.sum()
```

```python
# axis = 1 sums the columns for each row and returns a Pandas Series. This is what we want.

df.sum(axis=1)
```

```python
# assign Series back to DataFrame as overall_grade

df['overall_average'] = (df.sum(axis=1) / 3).astype('int')
```

```python
df.head()
```

### Optional... quick visualization of the distributioin of weight and speed values

```python
plt.hist(mammals.weight, bins=20, color='orange', edgecolor='black')
plt.title('Weight of Mammals in KG')
plt.xlabel('KG')
plt.ylabel('Count')

plt.show()
```

### What is the the weight of the fastest animal?

```python
# find the fastest speed using max()

(mammals.speed).max()
```

```python
# create a boolan Series or filter for observations that match the max speed

filt = mammals.speed == (mammals.speed).max()
filt
```

```python
# wrap the boolean Series in the original DataFrame to find observations that match the fastest speed

mammals[filt]
```

```python
# isolate the weight value if you want

mammals[filt]['weight']
```

### What is the overal percentage of specials?

```python
# we have a column of boolean values

mammals.specials
```

```python
# find the number of mammals classified as specials
# True boolean values are understood as having a value of 1, so you can sum the boolean Series

total_specials = mammals.specials.sum()
total_specials
```

```python
# find the number of mammals

total_mammals = len(mammals)
total_mammals
```

```python
print(f'{round(total_specials / total_mammals * 100, 2)}% of mammals in the df are specials.')
```

### How many animals are hoppers that are above the median speed? What percentage is this?

```python
# we can assign the subset of hoppers to a variable called hoppers if we like

hoppers = mammals[mammals.hoppers == True]
hoppers
```

```python
# find the median speed of all of the mammals

med_mammal_speed = mammals.speed.median()
med_mammal_speed
```

```python
# create a Series of boolean values for mammals in our hoppers subset that have a speed above the overall median speed

fast_hoppers = hoppers.speed > med_mammal_speed
fast_hoppers
```

```python
# wrap our Series of boolean values in our hoppers subset of the original DataFrame

hoppers[fast_hoppers]
```

```python
# we can assign our subset to a variable name to use in the next calculation

hoppers_above_median = hoppers[fast_hoppers]
```

```python
# find the percentage of hoppers that are above the median speed

round(len(hoppers_above_median) / len(hoppers) * 100)
```

```python
print(f'{round(len(hoppers_above_median) / len(hoppers) * 100)}% of hoppers in the mammals DataFrame have a speed above the overall median speed.')
```
