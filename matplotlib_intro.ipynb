```python
x1 = [student['n_ducks_owned'] for student in students if student['course'] == 'ds']
y1 = [student['avg_grade'] for student in students if student['course'] == 'ds']

x2 = [student['n_ducks_owned'] for student in students if student['course'] == 'wd']
y2 = [student['avg_grade'] for student in students if student['course'] == 'wd']

plt.scatter(x1, y1, label='Data Science')
plt.scatter(x2, y2, label='Web Development')
plt.legend()
```

```python
x1 = [student['coffee_consumption'] for student in students if student['course'] == 'ds']
y1 = [student['avg_grade'] for student in students if student['course'] == 'ds']

x2 = [student['coffee_consumption'] for student in students if student['course'] == 'wd']
y2 = [student['avg_grade'] for student in students if student['course'] == 'wd']

plt.figure(figsize=(16, 9))

plt.scatter(x1, y1, label='Data Science', alpha=.5)
plt.scatter(x2, y2, label='Web Development', alpha=.5)
plt.legend()
plt.title('Coffee Consumption vs Average Grade')
plt.xlabel('Coffee Consumed')
plt.ylabel('Average Grade')
```

```python
plt.figure(figsize=(16, 9))

plt.subplot(1, 2, 1)
plt.scatter(x1, y1)
plt.title('Data Science')
plt.xlabel('Coffee Consumption')
plt.ylabel('Average Grade')
plt.ylim(40, 120)
plt.xlim(-2, 4)

plt.subplot(1, 2, 2)
plt.scatter(x2, y2)
plt.title('Web Development')
plt.xlabel('Coffee Consumption')
plt.ylabel('Average Grade')
plt.ylim(40, 120)
plt.xlim(-2, 4)
```
