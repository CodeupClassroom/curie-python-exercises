# Plot Big O Notation with Matplotlib

### Directions
- Title your chart "Big O Notation"
- Label your x axis "Elements"
- Label your y axis "Operations"
- Label your curves or make a legend for the curves
- Use LaTex notation where possible

##### Curves to graph
- $y = 0x + 1$ and label the curve "O(1)"
- $y = log(x)$ and label the curve "O(log n)"
- $y = x$ and label the curve "O(n)"
- $y = x * log(x)$ and label it "O(n log n)"
- $y = x^2$ and label it "O(n^2)"
- $y = 2^x$ and label it "O(2^n)"
- $y = n!$ and label is "O(n!)"
- $y = n^n$ and label it "O(n^n)"

```python
import matplotlib.pyplot as plt
import math
```

```python
# inputs
x = range(1, 20)

# O(1)
y1 = [1 for n in x] 

# O(log n)
y2 = [math.log(n) for n in x] 

# O(n)
y3 = x 

# O(n log n)
y4 = [n * math.log(n) for n in x] 

# O(n^2)
y5 = [n**2 for n in x]

# O(2^n)
y6 = [2**n for n in x]

# y = n!
y7 = [math.factorial(n) for n in x]

# y = n^n
y8 = [n**n for n in x]
```

```python
plt.figure(figsize=(16, 12))
plt.title("Big O Notation", fontsize=20)

plt.xlabel("Number of Elements in a sequence", fontsize=16)
plt.ylabel("Number of Operations per element", fontsize=16)

plt.xlim(1, 11)
plt.ylim(0, 50)
plt.yscale("linear")

# O(1)
plt.plot(x, y1, color="darkseagreen", label="O(1)")

# O(log n)
plt.plot(x, y2, color="gold", label="O $log(x)$")

# O(n)
plt.plot(x, y3, color="palegreen", label="$O(x$)")

# O(n log n)
plt.plot(x, y4, color="skyblue", label="$O(n\ log\ n)$")

# O(n^2)
plt.plot(x, y5, color="blueviolet", label="$O(n^2)$")

# O(2^n)
plt.plot(x, y6, color="orchid", label="$O(2^n)$")

# y = n!
plt.plot(x, y7, color="crimson", label="$O(n!)$")

# y = n^n
plt.plot(x, y8, color="hotpink", label="$O(n^n)$")

# Setup the legend w/ a larger font size
plt.legend(loc='upper right', fontsize=16)

plt.show()
```
